**import modules**

```python
import numpy as np
import matplotlib.pyplot as plt
```

**data creation**

```python
np.random.seed(42)
x_train = 2 * np.random.rand(100)
y_train = 4 + 3 * x_train + np.random.randn(100)
```

```python
np.random.seed(20)
x_test = 2 * np.random.rand(100)
y_test = 4 + 3 * x_test + np.random.randn(100)
```

```python
x_train[0]
```

```python
plt.scatter(x_train,y_train)
plt.show()
```

**my_Linear_Regression class**

```python
class my_Linear_Regression():
  def __init__(self):
    self.slope=None
    self.intercept=None
  def fit(self,x_train,y_train):
    num=0
    den=0
    for i in range(x_train.shape[0]):
      num=num+((x_train[i]-x_train.mean())*(y_train[i]-y_train.mean()))
      den=den+((x_train[i]-x_train.mean())*(x_train[i]-x_train.mean()))
    self.slope=num/den
    self.intercept=y_train.mean()-(self.slope*x_train.mean())
  def predict(self,x_test):
    return self.intercept+self.slope*x_test
  def get_values(self):
    print("slope_value=>",self.slope)
    print("intercept_value=>",self.intercept)
  def mse(self, y_true, y_pred):
        return ((y_true - y_pred) ** 2).mean()
  def plot(self, x_train, y_train):
        plt.scatter(x_train, y_train, color="blue")
        x_start = x_train.min()
        x_end = x_train.max()
        y_start = self.intercept + (self.slope * x_start)
        y_end = self.intercept + (self.slope * x_end)
        plt.plot([x_start, x_end], [y_start, y_end], color="red")
        plt.show()

```

```python
model=my_Linear_Regression()
```

fit the model

```python
model.fit(x_train,y_train)
```

**prediction**


```python
model.predict(1)
```

**get_the_values**
slope and intercept


```python
model.get_values()
```

```python
model.mse(y_test,model.predict(x_test))
```
